```python
import pandas as pd
house=pd.read_csv('train.csv')
house.head()
```

```python
house=house.loc[:,house.dtypes!='object'].dropna(how='any')
y_train=house['SalePrice']
del house['SalePrice']
X_train=house
print(X_train.shape,y_train.shape)
```

```python
house=pd.read_csv('test.csv')
house.head()
```

```python
house=house.loc[:,house.dtypes!='object'].dropna(how='any')
X_test=house
print(X_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_test=lr.predict(X_test)
y_test
```

```python
data=list(zip(house['Id'],y_test))
data
```

```python
df = pd.DataFrame(data=data,columns=['Id','SalePrice'])
df
```

```python
df.to_csv('submission.csv',index=False)
```
