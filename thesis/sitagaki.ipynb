```python
%matplotlib inline

import pandas as pd

file = pd.read_csv('WEOApr2018all.csv')
file.shape
```

```python
file.head()
```

```python
gdp = file.loc[file['WEO Subject Code']=='NGDPD']
gdp.head()
```

```python
import missingno as msno
msno.matrix(df=gdp,figsize=(20,14),color=(0.5,0,0))
```

```python
target = gdp.loc[:,'2015']
print(target)
```

```python
for i in range(target.shape[0]): # 全ての数字を文字列に変換
        target.iloc[i]=str(target.iloc[i])
```

```python
for i in range(target.shape[0]): # 文字列中のカンマを削除
        target.iloc[i]=target.iloc[i].replace(",","")
print(target)
```

```python
import numpy as np
target = target.astype(np.float64) # 浮動小数に変換
print(target)
```

```python
target = target.fillna(target.median())
print(target)
```

```python
import seaborn as sns
ax = sns.distplot(target)
```
